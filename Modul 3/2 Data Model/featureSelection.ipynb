```python
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.api as sm

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.head()
```

```python
tips_dumy = pd.get_dummies(tips, drop_first=True, columns = ['sex','smoker','day','time'],dtype=int)
tips_dumy.head()
```

```python
tips_dumy.columns
```

```python
tips_reg = sm.OLS(tips_dumy['tip'],sm.add_constant(tips_dumy[['total_bill','size','sex_Female','smoker_No','day_Fri','day_Sat','day_Sun','time_Dinner']]))
hasil = tips_reg.fit()
print(hasil.summary())
```
