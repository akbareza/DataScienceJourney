# **INFERENCE STATISTICS : PARAMETRIC TEST**

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, ttest_ind, ttest_rel, f_oneway
```

```python
data = pd.read_csv("bestsellers with categories.csv")
data.head()
```

```python
data.info()
```

# **One Sample T-TEST**

Membandingkan mean dari suatu kelompok (kelompok A) dengan nilai mean teoritis yang diharapkan

- H0: Mean Harga buku = 10 dollar
- H1: Mean harga buku != 10 dollar

Significance level (alpha) = 0.05 = 5%

```python
data['Price'].mean() #belum valid karena hanya sample
```

```python
stat, p_val = ttest_1samp(data['Price'], 10, alternative='two-sided') #alternative is parameter to describe the operator of the H1 (alt hypothesis): two sided (!=), greater, less

#pengambilan kesimpulan bisa dari statistic value dan p value
#jika statistic value nanti akan dibandingkan dengan tabel statistiknya (T-table), cari yang Two-Sided degree of freedom nya 500 something
#jika PValue < aplha: Tolak H0, Pvalue > alpha, Gagal Tolak H0

#H0 adalah hipotesis yang menyatakan bahwa tidak ada perbedaan atau efek yang signifikan antara sampel dan populasi yang diperiksa.

if p_val <= 0.05:
    print(f'Kaerna P Value ({p_val}) <= 5%, maka Tolak H0\nDengan begini, kita punya cukup bukti untuk menolak H0, sehingga rata-rata Populasi harga buku != H0')
else:
    print(f'Kaerna P Value ({p_val}) > 5%, maka Gagal Tolak H0\nDengan begini, kita tidak punya cukup bukti untuk menolak H0, sehingga rata-rata Populasi harga buku = H0')
```

# **Independent Sample T-TEST**

Membandingkan mean dari suatu kelompok (kelompok A) dengan kelompok lainnya (kelompok B)

- H0: Mean harga buku fiksi >= Mean harga buku Non-Fiksi
- H1: Mean harga buku fiksi < Mean harga buku Non-Fiksi (ALTERNATIVE)

Significance level (alpha) = 0.01 = 1%

```python
data.groupby('Genre')[['Price']].mean().reset_index()
```

```python
dfic = data[data['Genre'] == 'Fiction']['Price']
dnfic = data[data['Genre'] == 'Non Fiction']['Price']
```

```python
stat_val, p_val = ttest_ind(dfic, dnfic, alternative="less")

if p_val <= 0.01:
    print(f'Kaerna P Value ({p_val}) <= 1%, maka Tolak H0\nDengan begini, kita punya cukup bukti untuk menolak H0, sehingga rata-rata Populasi harga buku fiksi >= Mean harga buku Non-Fiksi')
else:
    print(f'Kaerna P Value ({p_val}) > 1%, maka Gagal Tolak H0\nDengan begini, kita tidak punya cukup bukti untuk menolak H0, sehingga rata-rata Populasi harga buku fiksi < Mean harga buku Non-Fiksi')
```
