## **Inference**

1. Prove it 

- H0 Mean engagement video > 200?
- H1 Mean engagement viideo <= 200?

alpha = 0.5%

```python
#CLEANSING and converting column to numerical
data["Insights - Engagement"] = data["Insights - Engagement"].str.replace(',', '')
data["Insights - Engagement"] = pd.to_numeric(data["Insights - Engagement"])
```

```python
data["Insights - Engagement"].mean()
```

```python
stat_val, p_val = ttest_1samp(data[data["Content Type"]=='video']["Insights - Engagement"],200,alternative='less')
if p_val <= 0.05:
    print(f'Karena P Value ({p_val}) < 0.05, maka Tolak H0\nDengan begini, kita punya cukup bukti untuk menolak H0, sehingga rata-rata engagement video <= 200')
else:
    print(f'Karena P Value ({p_val}) > 0.05, maka Gagal Tolak H0\nDengan begini, kita tidak punya cukup bukti untuk menolak H0, sehingga rata-rata engagement video > 200')
```

2. Prove it 

- Apakah rata-rata reach dari caroulse != image?

alpha = 0.5%

```python
#CLEANSING and converting column to numerical
data["Insights - Reach"] = data["Insights - Reach"].str.replace(',', '')
data["Insights - Reach"] = pd.to_numeric(data["Insights - Reach"])
```

```python
data.groupby('Content Type')["Insights - Reach"].mean()
```

```python
a = data[data['Content Type']=='carousel']['Insights - Reach']
b = data[data['Content Type']=='image']['Insights - Reach']
stat_val, p_val = ttest_ind(a,b,alternative='two-sided')
f_p_value = '{:.20f}'.format(p_val)

if p_val <= 0.05:
    print(f'Karena P Value ({f_p_value}) < 0.05, maka Tolak H0\nDengan begini, kita punya cukup bukti untuk menolak H0, sehingga rata-rata reach dari caroulse = image')
else:
    print(f'Karena P Value ({f_p_value}) > 0.05, maka Gagal Tolak H0\nDengan begini, kita tidak punya cukup bukti untuk menolak H0, sehingga rata-rata reach dari caroulse != image')
```
