```python
# Library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Data

df = pd.read_csv('Sample_Superstore.csv',encoding='unicode_escape')
df.head()
```

```python
# Info 
df.info()
```

```python
# Descriptive Stat 

df.describe(include='all')
```

```python
# Cleaning Sementara (Remove Outliers)

df_clean = df[df['Profit'].between(-250, 250)]
df_clean.shape
```

# **1. LINE PLOT**
___

catatan:
1. Digunakan untuk merepresentasikan data **TIME SERIES** atau rentan waktu tertentu.
2. Sumbu X adalah data rentan waktu dan Y adalah nilai yang ingin ditampilkan
3. Bivariate (2 Variable): Numerical Vs Time Series

## **a. Data Time Series**

```python
#df_clean.info() #cek terlebih dahulu field time apakah sudah menjadi tipe datanya 'DATE'?

df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
df_clean['Ship Date'] = pd.to_datetime(df_clean['Ship Date'])
#df_clean.info()

df_clean['Order Month'] = df_clean['Order Date'].apply(lambda x : str(x)[:7])
df_clean['Order Month'] = pd.to_datetime(df_clean['Order Month'])
df_clean['Order Month Name'] = df_clean['Order Date'].dt.month_name()
display(df_clean.head()) #tanggal akan terisi otomatis di tanggal yang sama. hal ini dimaksudkan untuk memplotting sumbu x agar sama

df_cleanSales = df_clean.groupby('Order Month')['Sales'].mean().reset_index()
df_cleanSales.head()
```

```python
plt.figure(figsize=(15,5))


plt.plot(
    df_cleanSales['Order Month'],
    df_cleanSales['Sales']
)

plt.title('Rata rata penjualan setiap bulan (2016-2019)',size=15)
plt.xticks(df_cleanSales['Order Month'],rotation = 90)
plt.show()
```

## **c. Seaborn**

```python
plt.figure(figsize=(15,5))
sns.set_style('darkgrid')

sns.lineplot(
    data = df_clean,
    x = 'Order Month',
    y = 'Sales'
)
plt.title('Rata rata penjualan setiap bulan (2016-2019)',size=15)
plt.xticks(df_cleanSales['Order Month'],rotation = 90)
plt.show()
```

```python
# CEK UPPER LOWER 

print('Rumus confidence interval = Mean -+ 1.96 * Standard Error\n')

mean = df_clean[df_clean['Order Month'] == '2014-01-01' ]['Sales'].mean()
se = df_clean[df_clean['Order Month'] == '2014-01-01' ]['Sales'].sem() * 1.96
bound = (mean-se,mean+se)

print(f'Nilai Mean\t:{mean}')
print(f'Nilai SE\t:{se}')
print(f'Nilai Lower Bound\t:{bound[0]}')
print(f'Nilai Upper Bound\t:{bound[1]}')
```

```python
#BUAT 2 LINE PLOT

plt.figure(figsize=(15,5))
sns.set_style('darkgrid')

#LINE 1
sns.lineplot(
    data = df_clean[df_clean['Order Month Name']=='January'],
    x = 'Order Month',
    y = 'Sales',
    label = 'January'
)

#LINE 2
sns.lineplot(
    data = df_clean[df_clean['Order Month Name']=='February'],
    x = 'Order Month',
    y = 'Sales',
    label = 'February'
)


plt.title('Rata rata penjualan setiap bulan (2016-2019)',size=15)
plt.xticks(rotation = 90)
plt.show()
```

## **4. PIE CHART**
---
Catatan:
1. Menggambarkan proporsi/persentase dari sebuah kategori/kelompok.
2. Bivariate (Kategori Vs Numerical) biasanya berbentuk count.
3. Usahakan tidak menggunakan pie chart kecuali ingin menampilkan persentase.

```python
df_segment = df_clean.groupby('Segment')['Sales'].sum().reset_index()
display(df_segment)
df_segment['Percentage'] = df_segment['Sales']/df_segment['Sales'].sum() * 100
df_segment
```

```python
plt.figure(figsize=(10,5))

plt.pie(
    x = df_segment['Percentage'],
    labels= df_segment['Segment'],
    autopct='%.2f',
    explode=(0.001,0.05,0.5),
    shadow=True
)

plt.title('Proportion of Sales per Segement (%)')
plt.show()
```

## **5. HEAT MAP**
---
Catatan:
1. Warnan menunjukan intensitas dari nilai numerik
2. Biasanya semakin gelap warna, semakin besar nilainya.

```python
df_corr = df_clean[['Sales','Quantity','Discount','Profit']].corr()
df_corr
```

```python
plt.figure()
sns.heatmap(
    data = df_corr,
    annot = True,
    cmap = 'coolwarm'
    )
plt.show()
```

```python
matrix = np.triu(df_corr)

plt.figure()
sns.heatmap(
    data = df_corr,
    annot = True,
    cmap = 'coolwarm',
    mask=matrix
    )
plt.show()
```
