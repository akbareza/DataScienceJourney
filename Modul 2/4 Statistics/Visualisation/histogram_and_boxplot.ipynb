## **HISTOGRAM**

**Notes :**
1. Menampilkan Distribusi Data (Frekuensi atau Density)
2. Memperlihatkan frekuensi pada rentan nilai tertentu
3. Univariate (1 Variable) : Numerical (integer/float)
4. Library yang bisa digunakan : Matplotlib, Seaborn, Pandas, Plotly

```python
# Load Data

df = pd.read_csv('Sample_Superstore.csv',encoding='unicode_escape')
df.head()
```

```python
# Info 
df.info()
```

```python
# Descriptive Stat 

df.describe(include='all')
```

```python
# Cleaning Sementara (Remove Outliers)

df_clean = df[df['Profit'].between(-250, 250)]
df_clean.shape
```

### **A. MATPLOTLIB**

```python
# Ukuran Dari Kanvas / Ukuran Gambar
plt.figure(figsize=(10,5))

# Main Function Untuk Membuat Histogram
plt.hist(data = df, x = 'Profit', bins = 50, color='green')

# Label
plt.title('Profit Distribution', size = 20) # Judul
plt.xlabel('Profit', size = 15) # X Label
plt.ylabel('Frequency', size = 15) # Y Label

plt.legend(['Profit Distribution'])

plt.show()
```

```python
# Ukuran Dari Kanvas / Ukuran Gambar
plt.figure(figsize=(10,5))

# Main Function Untuk Membuat Histogram
plt.hist(data = df_clean, x = 'Profit', bins = 50, color='green')

# Label
plt.title('Profit Distribution', size = 20) # Judul
plt.xlabel('Profit', size = 15) # X Label
plt.ylabel('Frequency', size = 15) # Y Label

plt.legend(['Profit Distribution'])

plt.show()
```

### **SEABORN**

```python
## Cara 1

sns.displot(df_clean['Profit'], kde = True, aspect=2/1)
# KDE -> Biar ada garis distribusi
# Aspect -> Rasio Lebar/Tinggi
plt.title ('Distribusi Keuntungan Dari Toko')

plt.show()
```

```python
## Cara 2
plt.figure(figsize=(10,5))

sns.histplot(df_clean['Profit'], bins=20)
plt.title ('Distribusi Keuntungan Dari Toko')
plt.xticks(ticks=range(-250, 251, 50), rotation = 45)

plt.show()
```

```python
# Cara 3

sns.displot(data = df_clean, x = 'Profit', col = 'Segment')
plt.show()
```

```python
# Cara 3

sns.displot(data = df_clean, x = 'Profit', hue = 'Category')
plt.show()
```

```python
# Cara 4

sns.displot(data = df_clean, x = 'Profit', hue = 'Category', col = 'Segment')
plt.show()
```

```python
# KDE Plot (Kerbel Density Plot)

sns.kdeplot(
    data = df_clean,    # Data
    x = 'Profit',       # X yang digunakan
    hue = 'Segment',    # Dibagi Berdasarkan Apa
    fill = True,        # Membuat Area di Bawah Grafik Berwarna
    palette = 'tab10'   # Palette Warna
)

plt.show()
```
