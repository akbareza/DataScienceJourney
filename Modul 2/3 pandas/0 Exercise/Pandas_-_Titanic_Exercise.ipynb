**3. Passanger Alone or not ?**

Alone

```python
titanic['Alone'] = titanic['SibSp'] + titanic['Parch']
titanic.head()
```

```python
def aloneFunc(x):
    if x > 0:
        return 'not alone'
    else:
        return 'alone'

titanic['Alone'] = titanic['Alone'].apply(aloneFunc)
titanic.head()
```

**6. Calculate Surivivability**


Kemudian, coba lakukan analisis sederhana tentang `survivability` penumpang 

```python
titanic['Title'].value_counts()
```

```python
titanic[titanic['Title']=='the']
```

```python
def genderFunc(x):
    if x in ('Mr','Sir','Don'):
        return 'Male'
    elif x in ('Miss', 'Mrs', 'Ms','Mme','Mlle','Lady', 'the Countess'):
        return 'Female'
    else:
        return 'Male'

titanic['Gender'] = titanic['Title'].apply(genderFunc)
titanic['Gender'].value_counts()
```

```python
pivotRaw = titanic.pivot_table(values='Age',index='Gender',columns='Survived',aggfunc='count')
pivotRaw['surv_rate'] = pivotRaw[1]/ (pivotRaw[0]+pivotRaw[1])
pivotRaw
```
