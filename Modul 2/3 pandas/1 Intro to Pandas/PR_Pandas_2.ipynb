Di bawah ini adalah tampilan 5 baris pertama dari ketiga dataset tersebut

**Audi**

```python
df_audi = pd.read_csv('audi.csv')
df_audi.head()
```

**Hyundai**

```python
df_hyundai = pd.read_csv('hyundai.csv')
df_hyundai.head()
```

**Toyota**

```python
df_toyota = pd.read_csv('toyota.csv')
df_toyota.head()
```

## **No.1**

- Jika Anda perhatikan, nama kolom 'tax(£)' di dataset hyundai berbeda dengan di dataset lainnya.
- Ubah nama kolom tersebut menjadi 'tax' agar sama dengan di dataset yang lain.

```python
df_hyundai.columns.values[6] = 'tax'
df_hyundai
```

**Answer** ⬇️

```python
df_hyundai.rename(columns={'tax(£)': 'tax'},inplace=True)
df_hyundai.head(5)
```

## **No.2**

- Buat kolom baru bernama 'brand' dengan isi nama brand mobilnya, letakkan pada kolom paling kiri. 
- Misalnya:
    - Untuk data frame audi, isi kolom 'brand' dengan 'Audi'!
    - Untuk data frame hyundai, isi kolom 'brand' dengan 'Hyundai'!
    - Untuk data frame toyota, isi kolom 'brand' dengan 'Toyota'!


**Answer** ⬇️

```python
df_hyundai.insert(0,'brand','Hyundai')

df_audi.insert(0,'brand','Audi')

df_toyota.insert(0,'brand','Toyota')

display(df_hyundai.head(5),df_audi.head(5),df_toyota.head(5))
```

## **No.3**

- Gabungkan ketiga data frame tersebut menjadi sebuah data frame bernama df_car. Atur agar indexnya tidak ada yang berulang.


```python
df_car = pd.concat((df_audi,df_hyundai,df_toyota), ignore_index=True)
df_car
```

**Answer** ⬇️

```python
df_car = pd.concat((df_audi,df_hyundai,df_toyota)).reset_index().drop('index',axis=1)
df_car
```

## **No.4**

- Pada kolom 'model' di sebagian data, ada whitespace tambahan di awal atau di akhir yang tidak berguna. 
- Hapus whitespace tersebut dari semua data pada kolom 'model'!

**Answer** ⬇️

```python
df_car['model'] = df_car['model'].str.strip()
df_car
```

## **No.5**

- Dari df_car, tampilkan rata-rata pajak/tax berdasarkan brand dan transmission. Lalu urutkan dari tax yang paling tinggi.


```python
df_car[['brand','transmission','tax']].groupby(['brand','transmission']).mean().reset_index()
```

**Answer** ⬇️

```python
df_car.groupby(['brand','transmission'])[['tax']].mean().sort_values('tax', ascending=False).reset_index()
```

## **No.6**

- Anda diminta untuk mencari mobil Toyota Camry yang harganya (price) di bawah rata-rata harga Camry dan jarak tempuhnya (mileage) di bawah rata-rata mileage Camry juga. 
- Tampilkan mobil yang sesuai dengan kriteria tersebut!

**Answer** ⬇️

```python
camryPriceMean = df_car[df_car['model'].str.strip()=='Camry']['price'].mean()
camryMileageMean = df_car[df_car['model'].str.strip()=='Camry']['mileage'].mean()

df_car[(df_car['model'].str.strip()=='Camry') & (df_car['price']<camryPriceMean) & (df_car['mileage']<camryMileageMean)]
```

## **No.7**

Simpan df_car ke dalam format .csv dengan nama file 'all_cars.csv'!

**Answer** ⬇️

```python
df_car.to_csv('all_cars', index=False)
```

## **No.10**

```python
df_car[(df_car['brand']=='Toyota') & (df_car['year']==2019)][['model']].value_counts().sort_index()
```
