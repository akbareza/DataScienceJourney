```python
import numpy as np
import pandas as pd
```

# **MERGE AND JOIN**

```python
key = 'K0 K1 K2 K3'.split()
key2 = 'K0 K1 K3 K4'.split()
A = 'A0 A1 A2 A3'.split()
B = 'B0 B1 B2 B3'.split()
C = 'C0 C1 C2 C3'.split()
D = 'D0 D1 D2 D3'.split()

df1 = pd.DataFrame(data=list(zip(key,A,B)),columns=['key','A','B'])
df2 = pd.DataFrame(data=list(zip(key2,C,D)),columns=['key','C','D'])

display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how= 'inner', on='key')
```

```python
pd.merge(df1, df2, how= 'outer', on='key')
```

```python
pd.merge(df1, df2, how= 'left', on='key')
```

```python
pd.merge(df1, df2, how= 'right', on='key')
```

```python
df1.set_index('key',inplace=True)
df2.set_index('key',inplace=True)
df1
df2
```

```python
display(df1.join(df2,how='inner'))

display(df1.join(df2,how='outer'))

display(df1.join(df2,how='left'))

display(df1.join(df2,how='right'))

```

# **CONCAT**

```python
df1 = {
    'A' : 'A0 A1 A2 A3'.split(),
    'B' : 'B0 B1 B2 B3'.split(),
    'C' : 'C0 C1 C2 C3'.split(),
    'D' : 'D0 D1 D2 D3'.split(),
}

df2 = {
    'A' : 'A4 A5 A6 A7'.split(),
    'B' : 'B4 B5 B6 B7'.split(),
    'C' : 'C4 C5 C6 C7'.split(),
    'D' : 'D4 D5 D6 D7'.split(),
}

df3 = {
    'A' : 'A8 A9 A10 A11'.split(),
    'B' : 'B8 B9 B10 B11'.split(),
    'C' : 'C8 C9 C10 C11'.split(),
    'D' : 'D8 D9 D10 D11'.split(),
}

df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)

df4 = pd.DataFrame(np.random.randn(4,4),columns='W X Y Z'.split())
display(df1,df2,df3,df4)
```

```python
pd.concat([df1,df2,df3])
pd.concat([df1,df2,df3]).reset_index().drop('index', axis=1) # 0=baris, 1 = kolom
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
pd.concat([df1,df2,df3,df4],axis=0)
```
