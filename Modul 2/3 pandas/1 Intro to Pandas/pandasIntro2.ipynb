```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1,100,(5,5)), columns='A B C D E'.split())


df1 = pd.DataFrame({
    'A' : np.random.rand(5),
    'B' : np.random.randn(5),
    'C' : np.random.randint(1,100,5),
    'D' : np.linspace(1,100,5),
    'E' : np.arange(1,6)
})

display (df)
display (df1)
```

# **ADD NEW DATA**

```python
#Add Coloumn after the last coloumn
df['LastCol'] = np.random.randint(1,5,5)

#INSERT(pos, colname, value(list))
df.insert(2, "New Col", np.random.randint(1,5,5)) #randint menghasilkan tipe array
df
```

```python
display(df.iloc[1,4])
display(df.loc[1,'D'])
```

# DROP DATA

```python
display(df.drop('New Col', axis = 1)) #axis = 0 -> Row, axis = 1 -> Coloumn #This code just a temporary
display(df)
```

Inplace make the dataframe modified permanently

```python
df.drop('New Col', axis = 1, inplace=True) #axis = 0 -> Row, axis = 1 -> Coloumn #This code drop permanently
df
```

# INDEX ADV

```python
df2 = pd.DataFrame(np.random.randint(1,100,(5,5)), columns='A B C D E'.split(), index='Jakarta Bogor Surabaya Makassar Bali'.split())
df2
```

```python
df2['Gender'] = np.random.choice(('Male','Female'),5)
df2
```

```python
#RESET INDEX -> The current index will be moved to coloumn area
df2.reset_index(inplace=True,) #inplace for modified permanent
df2
```

```python
df2.set_index('index', inplace=True)
df2
```

```python
display(df2.sort_index(ascending=True)) #sot by index

display(df2.sort_values(['Gender','A'],ascending=[True,False])) #Sort by multicoloumn, inplace could be implemented to be modified permanently
```
